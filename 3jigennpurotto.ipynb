```python
import pandas as pd

file_path = '/content/drive/MyDrive/Colab Notebooks/hokkaido.csv' # Replace with your file path
try:
    df = pd.read_csv(file_path)
    display(df.head())
except FileNotFoundError:
    print(f"エラー: ファイルが {file_path} に見つかりませんでした。")
except Exception as e:
    print(f"エラーが発生しました: {e}")
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import japanize_matplotlib # Add this line to enable Japanese fonts

# Extract the data
x = df['comments_count']
y = df['like_count']
z = df['days_since_post']

# Create a new figure for the 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Create the 3D scatter plot
ax.scatter(x, y, z)

# Set labels for the axes with units
ax.set_xlabel('コメント数 (件)') # Added unit '件'
ax.set_ylabel('いいね数 (件)') # Added unit '件'
ax.set_zlabel('経過日数 (日)') # Added unit '日'

# Set the title of the plot
plt.title('北海度のコメント数、いいね数、経過日数の3次元プロット')

# Show the plot
plt.show()
```
